```python
def factorial (n):
    if n==1:
      return 1
    else:
       return n*factorial(n-1)

print(factorial(5))        
    
```

```python
def countdown(n):
    if n==0:
       return("Done")
    else:
        print (n)
        countdown(n-1)
    
countdown(6)    
```

```python
import numpy as np

numbers=[]
for i in range(5):
    num=int(input("Enter Marks:"))
    numbers.append(num)

    marks=np.array(numbers)

    
    max=np.max(marks)
    min=np.min(marks)
    avg=np.mean(marks)

    print("Highest:",max)
    print("Minimum:",min)
    print("Average:",avg)
```
