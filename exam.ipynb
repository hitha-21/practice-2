```python
def calculator():
    print("Simple Calculator")
    num1=int(input("Enter 1 number"))
    num2=int(input("Enter 2 number"))
    choice=input("Enter Choice")
    print("1.Add")
    print("2.Subtraction")
    print("3.Multiplication")
    print("4.Division")
    if choice=='1':
        print(f"\nResult={num1}+{num2}={num1+num2}")
    elif choice=='2':
         print(f"\nResult={num1}-{num2}={num1-num2}") 
    elif choice=='3': 
        print(f"\nResult={num1}x{num2}={num1*num2}") 
    elif choice=='4':
        if num2!=0:
            print(f"\nResult={num1}/{num2}={num1/num2}")
        else:
              print(f"Division with 0 not accepted")
    else:
        print(f"Invalid")
calculator() 
```

```python
def temp_converter():
    print("Temperature Converter")
    temp=float(input("Enter Temperature"))
    choice=input("Enter Choice")
    print("1.Celcius to Fahrenheit")
    print("2.Fahrenheit to Celcius")
    if choice=='1':
        fah=(temp*9/5+32)
        print(f"\nResult={temp}^c is {fah}^F")
    elif choice=='2':
        cel=(temp+32*5/9)
        print(f"\nResult={temp}^F is {cel}^C")
    else:
        print("Invalid")
temp_converter()            
```
